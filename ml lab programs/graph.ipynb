```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n = 200
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], n),
    'Value1': np.random.normal(50, 15, n),
    'Value2': np.random.normal(30, 10, n),
    'Group': np.random.choice(['X', 'Y'], n),
    'Score': np.random.randint(60, 100, n)
})
```

# Cell 1: Data Preparation and Bar Chart
This cell imports libraries, generates a synthetic dataset, and creates a bar chart. Below is an explanation of each line and the arguments used:
- `import numpy as np`: Imports NumPy for numerical operations and random number generation.
- `import pandas as pd`: Imports pandas for creating and handling tabular data (DataFrame).
- `import matplotlib.pyplot as plt`: Imports matplotlib for plotting charts.
- `import seaborn as sns`: Imports seaborn for advanced and aesthetic statistical plots.
- `np.random.seed(42)`: Sets the random seed for reproducibility, so results are the same every run.
- `n = 200`: Sets the number of samples in the dataset.
- `data = pd.DataFrame({...})`: Creates a DataFrame with columns:
    - `'Category'`: Randomly assigns one of four categories ('A', 'B', 'C', 'D') to each sample using `np.random.choice`.
    - `'Value1'`: Generates values from a normal distribution (mean=50, std=15) using `np.random.normal`.
    - `'Value2'`: Generates values from a normal distribution (mean=30, std=10) using `np.random.normal`.
    - `'Group'`: Randomly assigns one of two groups ('X', 'Y') using `np.random.choice`.
    - `'Score'`: Random integer scores between 60 and 99 using `np.random.randint`.
- `bar_data = data['Category'].value_counts()`: Counts frequency of each category for bar and pie charts.
- `plt.figure(figsize=(8, 5))`: Creates a new figure with width=8 and height=5 inches.
- `plt.bar(bar_data.index, bar_data.values, color='skyblue', edgecolor='black')`: Plots a bar chart:
    - `bar_data.index`: Category labels.
    - `bar_data.values`: Frequency of each category.
    - `color`: Fill color of bars.
    - `edgecolor`: Color of bar borders.
- `plt.title('Bar Chart- Category Counts')`: Sets the chart title.
- `plt.xlabel('Category')`: Sets the x-axis label.
- `plt.ylabel('Frequency')`: Sets the y-axis label.
- `plt.grid(axis='y', linestyle='--', alpha=0.6)`: Adds horizontal grid lines for readability.
    - `axis='y'`: Grid on y-axis only.
    - `linestyle='--'`: Dashed grid lines.
    - `alpha=0.6`: Grid line transparency.
- `plt.tight_layout()`: Adjusts layout to prevent overlap.
- `plt.show()`: Displays the plot.

### Why use `np.random.normal()` for data generation?

When creating sample data for visualizations, the choice of function depends on the pattern you want to demonstrate:
- **`np.random.normal()`** generates values that cluster around a mean, resembling real-world data (like heights or test scores) that naturally vary but center around an average. This is useful for showing natural variation and patterns in plots.
- **`np.random.randint()`** generates uniformly distributed values, where each value in the range is equally likely. This is good for showing data without a central tendency.

In this notebook, `np.random.normal()` is used to make the data look realistic and to help demonstrate how visualizations (like scatter plots and colorbars) reveal clusters and relationships. The normal distribution is not required for these plots, but it makes the visualization more meaningful by showing natural variation.

```python
bar_data = data['Category'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(bar_data.index, bar_data.values, color='skyblue', edgecolor='black')
plt.title('Bar Chart- Category Counts')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
bar_data = data['Category'].value_counts()
colors = sns.color_palette('pastel')
explode = [0.035] * len(bar_data)
print(explode)
plt.figure(figsize=(6, 6))
plt.pie(bar_data, labels=bar_data.index,autopct= '%.0f%%' ,startangle=90, colors=colors, explode=explode, shadow=True)
plt.title('Pie Chart- Category Distribution')
plt.tight_layout()
plt.show()
```

# Cell 2: Pie Chart of Category Distribution
This cell creates a pie chart to visualize the distribution of categories. Explanation of each line and arguments:
- `plt.figure(figsize=(6, 6))`: Creates a new figure with width=6 and height=6 inches (square for pie chart).
- `colors = sns.color_palette('pastel')`: Gets a pastel color palette from seaborn for pie slices.
- `explode = [0.05] * len(bar_data)`: Slightly separates each slice for emphasis (5% offset for each slice).
- `plt.pie(bar_data, labels=bar_data.index, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow=True)`: Plots the pie chart:
    - `bar_data`: Values for each slice (category counts).
    - `labels`: Names of each category.
    - `autopct`: Format for percentage labels on slices.
    - `startangle=90`: Rotates pie chart to start at 90 degrees.
    - `colors`: Color for each slice.
    - `explode`: Offset for each slice.
    - `shadow=True`: Adds shadow for visual effect.
- `plt.title('Pie Chart- Category Distribution')`: Sets the chart title.
- `plt.tight_layout()`: Adjusts layout to prevent overlap.
- `plt.show()`: Displays the plot.

### Why are two percent signs (`%%`) used at the end and one at the beginning in autopct format strings?
- In Python format strings, `%` is a special character used for formatting.
- The first `%` at the beginning (e.g., `%.1f`) indicates the format for the number (one decimal place).
- To display a literal percent sign after the number, you need to escape it by using `%%`.
- So, `'%.1f%%'` means:
    - `%.1f` formats the number to one decimal place.
    - `%%` displays a single percent sign in the output.
- Example: If the value is 23.5, `'%.1f%%'` will display `23.5%` in the pie chart slice.

### Why is `startangle` important in pie charts and why use 90?
- **`startangle`** sets the starting position (in degrees) for the first slice of the pie chart. By default, the first slice starts at the right (0 degrees).
- Setting `startangle=90` rotates the pie chart so the first slice starts at the top (12 o'clock position), which is visually intuitive and commonly used in publications and presentations.
- Using 90 degrees helps make the chart easier to read and compare, as viewers expect the first slice to begin at the top.

```python
plt.figure(figsize=(8, 5))
sns.set_palette('coolwarm')
sns.boxplot(data=data, x='Category', y='Value2', linewidth=2.5, fliersize=6, width=0.5)
plt.title('Boxplot with Custom Aesthetics')
plt.grid(True, axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# Cell 3: Box Plot of Value2 by Category
This cell creates a box plot to show the distribution of Value2 for each Category. Explanation of each line and arguments:
- `plt.figure(figsize=(8, 5))`: Creates a new figure with width=8 and height=5 inches.
- `sns.set_palette("coolwarm")`: Sets the color palette for seaborn plots to 'coolwarm'.
- `sns.boxplot(data=data, x='Category', y='Value2', linewidth=2.5, fliersize=6, width=0.5)`: Plots the box plot:

    - `x='Category'`: Categorical variable for grouping.
    - `y='Value2'`: Numerical variable to plot.
    - `linewidth=2.5`: Thickness of box edges.
    - `fliersize=6`: Size of outlier points.
    - `width=0.5`: Width of each box.
- `plt.title("Boxplot with Custom Aesthetics")`: Sets the chart title.
- `plt.grid(True, axis='y', linestyle='--')`: Adds horizontal grid lines for readability.

```python
plt.figure(figsize=(8, 5))
bins = np.arange(0, 100, 5)
for group in data['Group'].unique():
    plt.hist(data[data['Group'] == group]['Value1'], bins=bins, alpha=0.6, label=f'Group {group}')
plt.title('Histogram of Value1 by Group')
plt.xlabel('Value1')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

# Cell 5: Scatter Plot of Value1 vs Value2
This cell creates a scatter plot to show the relationship between Value1 and Value2, colored by Category and styled by Group. Explanation of each line and arguments:

    - `s=100`: Size of points.
    - `alpha=0.8`: Transparency of points.
    - `edgecolor='black'`: Color of point borders.
- `plt.title("Scatter Plot with Aesthetic Tweaks", fontsize=14, weight='bold')`: Sets the chart title with custom font size and weight.
