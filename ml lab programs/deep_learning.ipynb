```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# plt.scatter: creates a scatter plot of the data points
# x[:,0]: x-coordinates (feature 1) of the points
# x[:,1]: y-coordinates (feature 2) of the points
# c=y: colors each point according to its class label in y
# cmap=plt.cm.RdYlGn: uses the Red-Yellow-Green colormap for coloring points by class
# edgecolors='k': draws a black edge around each marker
# marker='o': uses circle markers for the points
# s=50: sets the marker size to 50
# plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.RdYlGn, edgecolors='k', marker='o', s=50)

# plt.colorbar: adds a colorbar to the plot to indicate the mapping of colors to class labels
# plt.colorbar(label="Class Label")

# Generate synthetic data
x,y = make_circles(n_samples=300, noise = 0.1, factor = 0.5, random_state=42)

# Plot the data
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=y, cmap=plt.cm.RdYlGn, marker='o', edgecolors='k', s=60)
plt.xlabel("Feature 1 [X:0]")
plt.ylabel("Feature 2 [X:1]")
plt.title("Plot of Points with Labels(make_circles)")
plt.colorbar(label="Class Label")
```
