```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

iris = load_iris()
x = iris.data
y = iris.target

#erros are dont put commas in True of heatmap and use capital letter in green word of second heatmap

multi_model = LogisticRegression( solver='lbfgs', max_iter=200)
multi_model.fit(x,y)
y_pred = multi_model.predict(x)

print("MultiClass Logistic Regression")
print(f"Accuracy Score : {accuracy_score(y,y_pred) * 100 : .2f}%")
print("C R")
print(classification_report(y, y_pred, target_names= iris.target_names))

cm_multi = confusion_matrix(y, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm_multi,cmap='Blues', annot=True, fmt='d',xticklabels= iris.target_names, yticklabels=iris.target_names)
plt.title("multiclass")
plt.xlabel("predicted")
plt.ylabel("actual")
plt.tight_layout()
plt.show()

binary_filter = y < 2
x_binary = x[binary_filter]
y_binary = y[binary_filter]

model = LogisticRegression(solver='lbfgs')
model.fit(x_binary, y_binary)
y_predb = model.predict(x_binary)

print("binary l r")
print(f"accuracy {accuracy_score(y_binary, y_predb) * 100 : .2f}%\n")
print(classification_report(y_binary, y_predb, target_names= iris.target_names[:2]))


cm_binary = confusion_matrix(y_binary, y_predb)
plt.figure(figsize=(8,6))
sns.heatmap(cm_binary,cmap='Greens', annot=True, fmt='d',xticklabels= iris.target_names[:2], yticklabels=iris.target_names[:2])
plt.title("binaryclass")
plt.xlabel("predicted")
plt.ylabel("actual")
plt.tight_layout()
plt.show()
```
