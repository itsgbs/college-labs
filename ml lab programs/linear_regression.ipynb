```python
# Tips for Practicing and Remembering ML Code, Arguments, and Functions
# ---------------------------------------------------------------
# 1. Practice by rewriting code from scratch and experimenting with variations.
# 2. Use official documentation and cheat sheets for reference.
#    - Official sources:
#      * Scikit-learn: https://scikit-learn.org/stable/documentation.html
#      * Pandas: https://pandas.pydata.org/docs/
#      * Numpy: https://numpy.org/doc/
#      * Matplotlib: https://matplotlib.org/stable/contents.html
#    - Cheat sheets:
#      * Scikit-learn cheat sheet: https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html
#      * Pandas cheat sheet: https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf
#      * Numpy cheat sheet: https://www.datacamp.com/cheat-sheet/numpy-cheat-sheet
#      * Matplotlib cheat sheet: https://matplotlib.org/cheatsheets/
# ---------------------------------------------------------------
# 3. Try small projects or Kaggle datasets to apply concepts.
# 4. Use auto-complete and tooltips in IDEs (like VS Code, Jupyter) to check arguments and functions.
# 5. Focus on understanding the logic, not memorizing everythingâ€”most programmers refer to docs often.
# 6. Make your own notes for frequently used functions and parameters.
# 7. Learn the most common libraries (pandas, numpy, matplotlib, sklearn) and their typical usage patterns.
# 8. With regular practice, you will naturally remember the most important names and arguments.
# ---------------------------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load the dataset
# You can replace this with: pd.read_csv('filename.csv') if reading from file
data = pd.DataFrame({
    'Hours': [1, 2, 3, 4.5, 5, 6, 7, 8.5, 9, 10],
    'Marks': [35, 40, 50, 60, 62, 70, 75, 85, 88, 95]
})

# Step 2: Separate the independent variable (X) and dependent variable (y)
X = data[['Hours']]  # Feature (2D array required by sklearn)
y = data['Marks']    # Target

# Step 3: Create and train the model
model = LinearRegression()  # Create a linear regression model
model.fit(X, y)             # Fit model to data

# Step 4: Make predictions
y_pred = model.predict(X)   # Predict using the trained model

# Step 5: Print model coefficients
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")

# Step 6: Evaluate model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")  # How well the regression line fits the data

# Step 7: Visualization
plt.figure(figsize=(10,10))
#set the size of the figure to 10x10 inches width and height
plt.scatter(X, y, color='blue', label='Actual Marks')      # Actual data points
plt.plot(X, y_pred, color='red', label='Regression Line')  # Regression line
plt.title('Study Hours vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Predict on full data and a test case
test_hours = pd.DataFrame({'Hours': [6.5]})
predicted_marks = model.predict(test_hours)
print(f"Predicted marks for 6.5 study hours: {predicted_marks[0]:.2f}")
```

    - `intercept_`: The y-intercept of the regression line.

- **Evaluate Model**:

## 2. Use of `plt.plot`

- Without it, only the actual data points are shown (scatter plot).
- The regression line helps visualize how well the model fits the data.
- If the line overlaps the points closely, removing it may not change the visual much, but it is still important for showing the model's fit.

## 3. Use of `plt.show()`

## 4. Why Underscore in `intercept_` and `coef_`
- In scikit-learn, attributes ending with `_` are set during model fitting (learned parameters).
- `intercept_`: y-intercept of the regression line.
- `coef_`: slope (coefficient) for each feature.

## 5. Why X is 2D and y is 1D
- `X` must be 2D (shape: [samples, features]) for sklearn models, even with one feature.
- `y` is 1D (just the target values for each sample).
- This format is required for sklearn to process the data correctly.

## 6. Why 0 Indexing in `coef_`
- `coef_` is an array because the model can have multiple features (inputs).
- For single feature regression, `coef_[0]` gives the slope for the first (and only) feature.
- For multiple features, you would access `coef_[1]`, `coef_[2]`, etc.

## 7. Variable Naming and Case Sensitivity

## 9. Why Different Slopes for Different Features and a Single Value for Intercept
- In linear regression, each feature (input variable) can have a different effect on the output. The slope for each feature (in `coef_`) shows how much the output changes when that feature increases by one unit, keeping other features constant.
- If you have multiple features, each gets its own slope value, so `coef_` is an array.
- The intercept (`intercept_`) is a single value because it represents the starting point of the regression line when all features are zero. It is not tied to any specific feature, but is a constant added to the prediction.
